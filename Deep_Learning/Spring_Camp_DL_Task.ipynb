# Play around with networks


```python
import numpy as np

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# A simple neuron
def neuron(x, weights, bias):
    z = np.dot(weights, x) + bias
    return sigmoid(z)

# Example input
x = np.array([0.5, 0.3])
weights = np.array([0.4, 0.7])
bias = -0.2

output = neuron(x, weights, bias)
print(f"Neuron output: {output}")
```

```python
import matplotlib.pyplot as plt

x_vals = np.linspace(-10, 10, 200)

def relu(x):
    return np.maximum(0, x)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, sigmoid(x_vals), label='Sigmoid')
plt.plot(x_vals, np.tanh(x_vals), label='Tanh')
plt.plot(x_vals, relu(x_vals), label='ReLU')
plt.title("Activation Functions")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Input data (2 features)
x = np.array([0.6, 0.9])

# Network architecture: 2 inputs → 2 hidden neurons → 1 output
# Initialize weights and biases
W1 = np.array([[0.2, 0.4],
               [0.5, -0.3]])  # shape (2, 2)
b1 = np.array([0.1, -0.2])   # shape (2,)

W2 = np.array([0.7, -0.5])   # shape (2,)
b2 = 0.3

# Forward pass
def forward(x):
    h = sigmoid(np.dot(W1, x) + b1)
    output = sigmoid(np.dot(W2, h) + b2)
    return output

output = forward(x)
print(f"Network output: {output}")
```

```python
# Mean Squared Error
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Binary Cross Entropy
def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-10  # to avoid log(0)
    return -np.mean(y_true * np.log(y_pred + epsilon) + (1 - y_true) * np.log(1 - y_pred + epsilon))

# Try both
y_true = np.array([1])
y_pred = np.array([0.85])

print("MSE Loss:", mse_loss(y_true, y_pred))
print("Binary Cross Entropy Loss:", binary_cross_entropy(y_true, y_pred))
```

# Forward Pass

```python
# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Forward pass function
def forward_pass(x, W1, b1, W2, b2):
    # TODO: Compute z1 (input to hidden layer)
    z1 = np.dot(W1, x) + b1

    # TODO: Apply sigmoid activation to z1
    h1 = sigmoid(z1)

    # TODO: Compute z2 (hidden to output layer)
    z2 = np.dot(W2, h1) + b2

    # TODO: Apply sigmoid to final output
    output = sigmoid(z2)

    return output

# Example: 3 inputs → 2 hidden neurons → 1 output neuron
x = np.array([0.4, 0.7, 0.1])
W1 = np.array([[0.2, -0.5, 0.3],
               [0.4, 0.1, -0.2]])
b1 = np.array([0.1, -0.3])

W2 = np.array([0.6, -0.4])
b2 = 0.2

# Run forward pass
output = forward_pass(x, W1, b1, W2, b2)
print(f"Network output: {output}")
```
